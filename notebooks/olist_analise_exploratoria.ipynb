```python
from pyspark.sql.functions import col, sum 

df_pedidos = spark.read.table("default.olist_pedidos")

df_pedidos.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in df_pedidos.columns
]).show()

df_pedidos.printSchema()
```
